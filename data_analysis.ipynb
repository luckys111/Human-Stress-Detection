```python

def comparisonplot(data, column1, column2, target):
    figure = px.scatter(data_frame = data, x=column1,
                    y=column2, color= target)
    # figure.show()

    return figure
```

```python
cols = []
figure=[]
for col1 in df.columns:
    cols.append(col1)
    for col2 in df.columns:
        if col2 not in cols:
            fig =comparisonplot(df,col1, col2, 'stress_level')
            figure.append(fig)
            # Save the figures

```
